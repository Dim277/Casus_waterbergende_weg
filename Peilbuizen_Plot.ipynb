```python
import pandas as pd
import matplotlib.pyplot as plt

# Lees CSV-bestand in met puntkomma als scheidingsteken voor het tweede bestand (PB02)
df_PB02 = pd.read_csv('PB02_01122020_01122023.csv', skipinitialspace=True, header=0, sep=';')

# Maak een datum kolom voor het tweede DataFrame
df_PB02['DateTime'] = pd.to_datetime(df_PB02['result_timestamp'], format='%d-%m-%Y %H:%M')

# Toon de kolomnamen
#print(df_PB02.columns)
#display(df_PB02)
#df_PB02.head(20)

# Plot de DateTime tegenover het water_level
plt.figure(figsize=(20, 5))
plt.plot(df_PB02["DateTime"], df_PB02["water_level"], label="Waterstand (m)")
plt.title("Waterstand over tijd")
plt.xlabel("Datum")
plt.ylabel("Waterstand [m]")
plt.grid(True)
plt.legend()
plt.show()
```
