```python
import pandas as pd
import matplotlib.pyplot as plt

# Lees CSV-bestand in met puntkomma als scheidingsteken
df_PB02 = pd.read_csv('PB02_01122020_01122023.csv', skipinitialspace=True, header=0, sep=';')
df_PB03 = pd.read_csv('PB03_01122020_01122023.csv', skipinitialspace=True, header=0, sep=';')
df_PB07 = pd.read_csv('PB07_01112022_01122023.csv', skipinitialspace=True, header=0, sep=';')

# Maak een datum kolom voor PB02, PB03 en PB07
df_PB02['DateTime'] = pd.to_datetime(df_PB02['result_timestamp'], format='%d-%m-%Y %H:%M')
df_PB03['DateTime'] = pd.to_datetime(df_PB03['result_timestamp'], format='%d-%m-%Y %H:%M')
df_PB07['DateTime'] = pd.to_datetime(df_PB07['result_timestamp'], format='%d-%m-%Y %H:%M')

# Plot de DateTime tegenover het water_level voor PB
plt.figure(figsize=(20, 5))
plt.plot(df_PB02["DateTime"], df_PB02["water_level"], label="PB02 - Waterstand (mNAP)")
plt.plot(df_PB03["DateTime"], df_PB03["water_level"], label="PB03 - Waterstand (mNAP)")
plt.plot(df_PB07["DateTime"], df_PB07["water_level"], label="PB07 - Waterstand (mNAP)")

# Opmaak van de grafiek
plt.title("Waterstand over tijd [PB02, PB03, PB07]")
plt.xlabel("Datum")
plt.ylabel("Waterstand [mNAP]")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lees CSV-bestand in met puntkomma als scheidingsteken voor PB
df_PB02 = pd.read_csv('PB02_01122020_01122023.csv', skipinitialspace=True, header=0, sep=';')
df_PB03 = pd.read_csv('PB03_01122020_01122023.csv', skipinitialspace=True, header=0, sep=';')
df_PB07 = pd.read_csv('PB07_01112022_01122023.csv', skipinitialspace=True, header=0, sep=';')

# Maak een datum kolom voor PB
df_PB02['DateTime'] = pd.to_datetime(df_PB02['result_timestamp'], format='%d-%m-%Y %H:%M')
df_PB03['DateTime'] = pd.to_datetime(df_PB03['result_timestamp'], format='%d-%m-%Y %H:%M')
df_PB07['DateTime'] = pd.to_datetime(df_PB07['result_timestamp'], format='%d-%m-%Y %H:%M')

# Filter de data op het tijdsinterval 2023-08-01 tot 2023-08-30
start_date = '2023-08-01'
end_date = '2023-08-30'

# Filter PB02
mask_PB02 = (df_PB02['DateTime'] >= start_date) & (df_PB02['DateTime'] <= end_date)
df_PB02_filtered = df_PB02.loc[mask_PB02]

# Filter PB03
mask_PB03 = (df_PB03['DateTime'] >= start_date) & (df_PB03['DateTime'] <= end_date)
df_PB03_filtered = df_PB03.loc[mask_PB03]

# Filter PB07
mask_PB07 = (df_PB07['DateTime'] >= start_date) & (df_PB07['DateTime'] <= end_date)
df_PB07_filtered = df_PB07.loc[mask_PB07]

# Plot de DateTime tegenover het water_level voor PB
plt.figure(figsize=(20, 5))
plt.plot(df_PB02_filtered["DateTime"], df_PB02_filtered["water_level"], label="PB02 - Waterstand (mNAP)")
plt.plot(df_PB03_filtered["DateTime"], df_PB03_filtered["water_level"], label="PB03 - Waterstand (mNAP)")
plt.plot(df_PB07_filtered["DateTime"], df_PB07_filtered["water_level"], label="PB07 - Waterstand (mNAP)")

# Opmaak van de grafiek
plt.title("Waterstand over tijd [PB02, PB03, PB07] - Tijdsinterval: {} tot {}".format(start_date, end_date))
plt.xlabel("Datum")
plt.ylabel("Waterstand [mNAP]")
plt.grid(True)
plt.legend()
plt.show()
```
